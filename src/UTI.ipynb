```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('CLA.csv')

columns = ['CLAA07', 'CLAA08', 'CLAA09b', 'CLAA10']
names = ['Gonorrheal screen','Chlamydial screen','Syphilis test','HIV test']
print("1	Positive \n2	Negative \n3	Contaminated \nD	Don't know\n")
for column,name in zip(columns,names):
    value_counts = df[column].value_counts()
    print(f"Value counts for {name}:")
    print(value_counts)
    print()
```
